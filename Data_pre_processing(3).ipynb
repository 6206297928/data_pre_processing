```python
# Total missing values per column
print(df.isna().sum())

# Total missing values in entire DataFrame
print(df.isna().sum().sum())
```

```python
missing_columns = df.columns[df.isna().any()]
print("Columns with missing values:", missing_columns)
```

```python
def count_nan_non_nan(df, column):
    nan_count = df[column].isna().sum()
    non_nan_count = df[column].notna().sum()
    return nan_count, non_nan_count

nan_count, non_nan_count = count_nan_non_nan(df, "Flow Bytes/s")
print(f"NaN values: {nan_count}")
print(f"Non-NaN values: {non_nan_count}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size dynamically
plt.figure(figsize=(10,6))

# Histogram with KDE
sns.histplot(df['Flow Bytes/s'], bins=50, color='skyblue', kde=True)

# Add grid
plt.grid(True, linestyle='--', alpha=0.5)

# Labels and title
plt.xlabel('Flow Bytes/s')
plt.ylabel('Frequency')
plt.title('Distribution of Flow Bytes/s with KDE')

plt.show()
```

```python
df["Flow Bytes/s"] = pd.to_numeric(df["Flow Bytes/s"], errors='coerce')

# Now fill NaNs with median
df["Flow Bytes/s"] = df["Flow Bytes/s"].fillna(df["Flow Bytes/s"].median())
```

```python
print(df.isna().sum())

# Total missing values in entire DataFrame
print(df.isna().sum().sum())
```

```python
def count_nan_non_nan(df, column):
    nan_count = df[column].isna().sum()
    non_nan_count = df[column].notna().sum()
    return nan_count, non_nan_count

nan_count, non_nan_count = count_nan_non_nan(df, "Flow Bytes/s")
print(f"NaN values: {nan_count}")
print(f"Non-NaN values: {non_nan_count}")
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x='Label', data=df)

# Custom tick names for bins
ax.set_xticklabels(['BENIGN', 'ATTACK'])

plt.title('Class Distribution')
plt.xlabel('Traffic Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Label'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue','salmon'],
    explode=(0,0.1),
    labels=['BENIGN','ATTACK']
)
plt.title("Class Distribution (Pie Chart)")
plt.ylabel("")  # hide y-label
plt.show()
```
